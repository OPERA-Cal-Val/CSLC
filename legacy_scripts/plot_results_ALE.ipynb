```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
# Read to pandas dataframe and rename columns
raw = pd.read_csv(f'ALE.csv', index_col=0, names=['Date', 'ALE_Rg_Mean', 'ALE_Rg_Stdev', 'ALE_Az_Mean', 'ALE_Az_Stdev'])
deramped = pd.read_csv(f'ALE_withDeramp.csv', index_col=0, names=['Date', 'ALE_Rg_Mean', 'ALE_Rg_Stdev', 'ALE_Az_Mean', 'ALE_Az_Stdev'])

# Convert to datetime format
dat = pd.to_datetime(raw.Date, format='%Y%m%d')
```

```python
#Azimuth
fig, ax = plt.subplots(figsize=(8,6))
ax.set_ylim(-20,20)
ax.errorbar(dat, raw.ALE_Az_Mean, yerr = raw.ALE_Az_Stdev, fmt='o', alpha = 0.8, label='raw')
ax.errorbar(dat, deramped.ALE_Az_Mean, yerr = deramped.ALE_Az_Stdev, fmt='o', alpha = 0.8, label='deramped')

ax.set_title(f'Northing')
ax.set_xlabel('Date')
ax.set_ylabel('$\Delta Y$ (m)')
plt.legend()
fig.suptitle(f'Absolute location error for t064_135523_iw2 (Rosamond)')
fig.savefig(f'ALE_geoSLC_dNorthing.png',dpi=300,bbox_inches='tight')
```

```python
#Range
fig, ax = plt.subplots(figsize=(8,6))
ax.set_ylim(-20,20)
ax.errorbar(dat, raw.ALE_Rg_Mean, yerr = raw.ALE_Rg_Stdev, fmt='o', alpha = 0.8, label='raw')
ax.errorbar(dat, deramped.ALE_Rg_Mean, yerr = deramped.ALE_Rg_Stdev, fmt='o', alpha = 0.8, label='deramped')

ax.set_title(f'Easting')
ax.set_xlabel('Date')
ax.set_ylabel('$\Delta X$ (m)')
plt.legend()
fig.suptitle(f'Absolute location error for t064_135523_iw2 (Rosamond)')
fig.savefig(f'ALE_geoSLC_dEasting.png',dpi=300,bbox_inches='tight')
```
